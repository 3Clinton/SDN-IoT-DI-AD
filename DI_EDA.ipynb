```python
import pandas as pd
import random


# Dataset statistics (class names and instance counts)
class_counts = {
    "Hudp_flood.pcap": 5353295,
    "hmqtt_TcpSyn_flood.pcap": 4816624,
    "Hicmp_flood.pcap": 4572914,
    "Htcp_flood.pcap": 4426750,
    "Btcp_flood.pcap": 3901461,
    "Bmqtt_udp_flood.pcap": 3466258,
    "Bhttp_flood.pcap": 3272116,
    "Budp_flood.pcap": 3228636,
    "mqtt_publish_flood1.pcap": 174881,
    "normal": 89973,  # Target count for balancing
    "mqtt_publish_flood.pcap": 39406,
    "Bicmp_flood.pcap": 14602,
}

# Load the dataset

# Separate the data by class
balanced_df = pd.DataFrame()  # To store balanced data

for class_name, count in class_counts.items():
    class_subset = combined_df[combined_df["type"] == class_name]  # Filter by class
    
    if len(class_subset) > class_counts["normal"]:  # If more than 89,973 instances
        class_subset = class_subset.sample(class_counts["normal"], random_state=42)  # Undersample
    elif len(class_subset) < class_counts["normal"]:  # If fewer, keep as is
        class_subset = class_subset  # No oversampling applied here
    
    balanced_df = pd.concat([balanced_df, class_subset])  # Merge to final dataset

# Save the balanced dataset
balanced_df.to_csv("balanced_dataset.csv", index=False)

print("[INFO] Dataset balanced successfully! New class distribution:")
print(balanced_df["type"].value_counts())
```
